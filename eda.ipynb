```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel("GuttmacherInstituteAbortionDataByState.xlsx")
df.head()
```

```python
df.columns
```

```python
df['Clinic Change'] = pd.to_numeric(df['% change in the no. of abortion clinics, 2017-2020'], errors='coerce')
df['Access Gap'] = pd.to_numeric(df['% of women aged 15-44 living in a county without a clinic, 2020'], errors='coerce')
```

```python
import plotly.graph_objects as go

# Sort states by access gap
df_sorted = df.sort_values('Access Gap', ascending=True)

fig3 = go.Figure()

# Clinic Change line
fig3.add_trace(go.Scatter(
    x=df_sorted['U.S. State'],
    y=df_sorted['% of residents obtaining abortions who traveled out of state for care, 2020'],
    mode='lines+markers',
    name='Clinic Change',
    line=dict(color='green'),
    yaxis='y1'
))

# Access Gap area
fig3.add_trace(go.Bar(
    x=df_sorted['U.S. State'],
    y=df_sorted['Access Gap'],
    name='Access Gap',
    marker_color='red',
    opacity=0.6,
    yaxis='y2'
))

fig3.update_layout(
    title="Clinic Change vs. Access Gap by State",
    xaxis=dict(title='State', tickangle=-45),
    yaxis=dict(title='Clinic Change (%)', side='left'),
    yaxis2=dict(title='% Women Without Clinic', overlaying='y', side='right'),
    height=800,
    barmode='overlay'
)

fig3.show()
```

```python
import plotly.graph_objects as go

# Sort states by Access Gap
df_sorted = df.sort_values('Access Gap', ascending=True)

# Flip one metric (so bars go opposite directions)
travel = df_sorted['% of residents obtaining abortions who traveled out of state for care, 2020']
access_gap = -df_sorted['Access Gap']

fig_div = go.Figure()

fig_div.add_trace(go.Bar(
    x=access_gap,
    y=df_sorted['U.S. State'],
    name='Access Gap',
    orientation='h',
    marker_color='red'
))

fig_div.add_trace(go.Bar(
    x=travel,
    y=df_sorted['U.S. State'],
    name='% Traveled Out of State (2020)',
    orientation='h',
    marker_color='green'
))

fig_div.update_layout(
    title="Divergent Bar: Access Gap vs Out-of-State Travel by State",
    xaxis=dict(title='Percentage (Negative = Access Gap, Positive = Travel)', zeroline=True),
    yaxis=dict(title='State'),
    barmode='relative',  # required for divergent layout
    height=800
)

fig_div.show()
```
